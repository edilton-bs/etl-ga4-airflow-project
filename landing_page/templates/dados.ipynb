```python
import requests
import csv
from time import sleep

BASE_URL = 'https://api.themoviedb.org/3'
API_KEY = 'ad4fb9197822289cfcb5b3b145474b3f'

def get_director(credits):
    for crew_member in credits.get('crew', []):
        if crew_member.get('job') == 'Director':
            return crew_member.get('name')
    return None

def has_trailer(videos):
    for video in videos.get('results', []):
        if video.get('type') == 'Trailer' and video.get('site') in ['YouTube', 'Vimeo']:
            return True
    return False

def get_trailer_url(videos):
    for video in videos.get('results', []):
        if video.get('type') == 'Trailer' and video.get('site') == 'YouTube':
            return f"https://www.youtube.com/watch?v={video.get('key')}"
    return None

results = []
for decade_start in range(1960, 2030, 5):
    decade_end = min(decade_start + 4, 2025)
    page = 1
    collected = 0
    while collected < 4000:
        discover_url = (
            f"{BASE_URL}/discover/movie?api_key={API_KEY}&language=pt-BR"
            f"&sort_by=popularity.desc"
            f"&vote_average.gte=3"
            f"&vote_count.gte=700"
            f"&primary_release_date.gte={decade_start}-01-01"
            f"&primary_release_date.lte={decade_end}-12-31"
            f"&page={page}"
        )
        discover_resp = requests.get(discover_url).json()
        movies = discover_resp.get('results', [])
        if not movies:
            break
        for movie in movies:
            movie_id = movie['id']
            # Detalhes + créditos + vídeos
            details_url = (
                f"{BASE_URL}/movie/{movie_id}?api_key={API_KEY}&language=pt-BR&append_to_response=credits,videos"
            )
            details = requests.get(details_url).json()
            # Verifica se tem trailer
            if not has_trailer(details.get('videos', {})):
                continue
            # Diretor
            diretor = get_director(details.get('credits', {}))
            if not diretor:
                continue
            # Monta dados
            overview = (details.get('overview') or '')[:2100]
            poster_path = details.get('poster_path')
            poster_url = f"https://image.tmdb.org/t/p/w500{poster_path}" if poster_path else ''
            results.append({
                'Nome do filme': details.get('title'),
                'Diretor': diretor,
                'Avaliação': details.get('vote_average'),
                'Gênero': details.get('genres', [{}])[0].get('name'),
                'Ano de lançamento': details.get('release_date', '')[:4],
                'Língua': details.get('original_language'),
                'Sinopse': overview,
                'Link para a capa': poster_url,
                'Trailer': get_trailer_url(details.get('videos', {}))
            })
            collected += 1
            if collected >= 4000:
                break
        page += 1
        sleep(0.25)  # Evita limite de requisições


# criar o dataframe
df = pd.DataFrame(results)
```
